```python
rs = pd.read_csv('C:/Users/Hitesh Sachdeva/Downloads/6 sem/dataset/ratings.csv')
mv = pd.read_csv('C:/Users/Hitesh Sachdeva/Downloads/6 sem/dataset/movies.csv')
```

```python
rs = pd.merge(mv,rs).drop(['genres','timestamp'],axis=1)
rs.head()
```

```python
u_rs = rs.pivot_table(index=['userId'],columns=['title'],values='rating')
u_rs.head()
```

```python
u_rs = u_rs.dropna(thresh=10,axis=1).fillna(0)
u_rs.head()
```

```python
no_user_voted = rs.groupby('movieId')['rating'].agg('count')
no_movies_voted = rs.groupby('userId')['rating'].agg('count')
```

```python
f,ax = plt.subplots(1,1,figsize=(16,4))
plt.scatter(no_user_voted.index,no_user_voted,color='purple')
plt.axhline(y=10,color='r')
plt.xlabel('MovieId')
plt.ylabel('No. of users voted')
plt.show()
```

```python
f,ax = plt.subplots(1,1,figsize=(16,4))
plt.scatter(no_movies_voted.index,no_movies_voted,color='orange')
plt.axhline(y=50,color='g')
plt.xlabel('UserId')
plt.ylabel('No. of votes by user')
plt.show()
```

```python
item_similarity_data = u_rs.corr(method='pearson')
item_similarity_data.head(50)
```

```python
def get_similar_movies_from_dataset(movie_name,user_rating):
    similar_score = item_similarity_data[movie_name]*(user_rating-2.5)
    similar_score = similar_score.sort_values(ascending=False)

    return similar_score
```

```python
user1 = [("Jumanji (1995)",5),
("12 Years a Slave (2013)",4),
("2 Fast 2 Furious (Fast and the Furious 2, The) (2003)",3)]

similar_movies = pd.DataFrame()

for movie,rating in user1:
    similar_movies = similar_movies.append(get_similar_movies_from_dataset(movie,rating),ignore_index=True)
    
similar_movies.head(10)
similar_movies.sum().sort_values(ascending=False)
```

```python
user2 = [("2012 (2009)",3),
("Nixon (1995)",4),
("American President, The (1995)",5)]

a_similar_movies = pd.DataFrame()

for movie,rating in user2:
    a_similar_movies = a_similar_movies.append(get_similar_movies_from_dataset(movie,rating),ignore_index=True)
    
a_similar_movies.head(5)
a_similar_movies.sum().sort_values(ascending=False)
```
